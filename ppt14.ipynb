```python
3
class ListNode:
    def __init__(self, val=0, next=None, bottom=None):
        self.val = val
        self.next = next
        self.bottom = bottom

def mergeLists(head1, head2):
    dummy = ListNode()  # Dummy node to handle edge cases
    tail = dummy

    while head1 and head2:
        if head1.val < head2.val:
            tail.bottom = head1
            head1 = head1.bottom
        else:
            tail.bottom = head2
            head2 = head2.bottom
        tail = tail.bottom

    # Append the remaining nodes of either sub-linked list
    if head1:
        tail.bottom = head1
    else:
        tail.bottom = head2

    return dummy.bottom

def flattenLinkedList(head):
    if not head or not head.next:
        return head

    head.next = flattenLinkedList(head.next)  # Recursively flatten the rest of the list
    head = mergeLists(head, head.next)  # Merge the current list with the flattened rest

    return head
```
