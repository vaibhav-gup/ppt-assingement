```python
1
def closest(arr, target):
    arr.sort()  # Sorting the array in ascending order
    closest_sum = float('inf')  # Variable to store the closest sum
    n = len(arr)  # Length of the array

    # Loop through the array, up to the third-to-last element
    for i in range(n - 2):
        left = i + 1  # Left pointer
        right = n - 1  # Right pointer

        # Perform two-pointer approach
        while left < right:
            current_sum = arr[i] + arr[left] + arr[right]  # Calculate the current sum

            # Check if the current sum is closer to the target than the previous closest sum
            if abs(current_sum - target) < abs(closest_sum - target):
                closest_sum = current_sum

            # Adjust the pointers based on the current sum compared to the target
            if current_sum < target:
                left += 1
            elif current_sum > target:
                right -= 1
            else:
                return target  # If the current sum is equal to the target, return the target sum

    return closest_sum  # Return the closest sum to the target

            
            
                    
                
          space complexity is O(1) and the time complexity is O(n^2)   
        
```

```python
3
def lexo(arr):
    c = 0  # Variable to count the number of swaps
    n = len(arr)  # Length of the array
    arr.sort()  # Sort the array in ascending order
    
    # Iterate from the last element to the first element
    for i in range(n - 1, -1, -1):
        if arr[n - 1 - i] > arr[i]:
            # Swap the elements if the left element is greater than the right element
            t = arr[n - 1 - i]
            arr[n - 1 - i] = arr[i]
            arr[i] = t
            c += 1  # Increment the swap count
        else:
            break  # Exit the loop if the elements are already in lexicographically increasing order
    
    return arr, c  # Return the sorted array and the swap count


        
        
     space complexity is O(1) and the time complexity is O(nlogn) f   
```

```python
5 
def oneplus(digit):
    n = len(digit)  # Length of the input digit array
    
    # Iterate from the last digit to the first digit
    for i in range(n - 1, -1, -1):
        if digit[i] < 9:
            # If the current digit is less than 9, increment it by 1 and return the updated digit array
            digit[i] += 1
            return digit
        else:
            digit[i] = 0  # Set the current digit to 0 since it becomes 10 after incrementing
    
    # If all digits are 9, add 1 as the most significant digit
    return [1] + digit

   The time complexity of this code is O(n) 
    The space complexity is also O(n) 
```

```python
6
def singleOne(nums):
    n = len(nums)  # Length of the input nums array
    c = 0  # Counter variable to count the occurrences of 1
    
    for i in range(n):
        if nums[i] == 1:
            c += 1  # Increment the counter if the current element is 1
        
        # Check if there are multiple occurrences of 1
        elif c >= 2:
            return "Error"  # Return "Error" if there are two or more occurrences of 1
        
        # Check if there is only one occurrence of 1
        if c == 1:
            return 1  # Return 1 if there is only one occurrence of 1 in the array
    
    # Return "Error" if there are no occurrences of 1 or multiple occurrences of 1
    return "Error"


The space complexity is also O(1)
The space complexity is also O(1)
```

```python
7
def findMissingRanges(nums, lower, upper):
    result = []
    start = lower

    for num in nums:
        if num >= start:
            if num == start:
                start += 1
            else:
                result.append([start, num - 1])
                start = num + 1

    if start <= upper:
        result.append([start, upper])

    return result


#Initialize an empty list to store the ranges.
#Set a variable start to lower.
#Iterate through each number num in the sorted nums array:
#If num is equal to or greater than start, update start to num + 1 and continue to the next iteration.
#If num is less than start, continue to the next iteration.
#If num is greater than start + 1, there is a missing range. Append the range [start, num - 1] to the result list and update start to num + 1.
#After the loop, if start is less than or equal to upper, there is a missing range from start to upper. Append the range [start, upper] to the result list.
#Return the result list containing all the ranges.

time complexity is O(n) and the space complexity is O(1).
```

```python
8
def getStartTime(interval):
    return interval[0]

def canAttendMeetings(intervals):
    intervals.sort(key=getStartTime)  # Sort intervals based on start time
    
    for i in range(1, len(intervals)):
        if intervals[i][0] < intervals[i-1][1]:
            return False  # Overlap found, return False
    
    return True  # No overlaps found, return True

time complexity is O(nlogn) and the space complexity is O(1)
```
